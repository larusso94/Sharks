```python
from fastai.vision.all import *
from fastai.vision.widgets import *
```


# The Amazing Bear Classifier! 
You need to know whether you're being chased by a dangerous white shark, or a cute whale shark, and you need an answer fast? Then you've come to the right place. Take a pic of the potentially vicious killer, and click 'upload' to classify it. (Important: this only handles white sharks, bull sharks, mako sharks, whale sharks, basking sharks, thresher sharks, tiger sharks, whitetip sharks, hammerhead sharks, blue sharks, nurse sharks, blacktip sharks, sand tiger sharks, lemon sharks. It will not give a sensible answer for other species of sharks nor other things that are not sharks.

```python
path = Path()
learn_inf = load_learner(path/'export.pkl', cpu=True)
btn_upload = widgets.FileUpload()
out_pl = widgets.Output()
lbl_pred = widgets.Label()
```

```python
def on_data_change(change):
    lbl_pred.value = ''
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))
    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'
```

```python
btn_upload.observe(on_data_change, names=['data'])
```

```python
display(VBox([widgets.Label('Select your shark!'), btn_upload, out_pl, lbl_pred]))
```
